```python
import os
from Bio import SeqIO
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from pandas import DataFrame
from sklearn.cluster import KMeans
```

```python
notebook_path = os.path.abspath("genomeClustering.ipynb")
path = os.path.join(os.path.dirname(notebook_path), "data/J29B_expected.fasta")
pathFASTQ= os.path.join(os.path.dirname(notebook_path), "data/fastq/J29_B_CE_IonXpress_005.fastq")

expected = list(SeqIO.parse(path, "fasta"))
    
from Bio import Align
aligner = Align.PairwiseAligner()
aligner.mode = 'global'

aligner.match_score = 1
aligner.mismatch_score = -1
aligner.open_gap_score = -10
aligner.extend_gap_score = -10
aligner.target_end_gap_score = -1
aligner.query_end_gap_score = -1

#print(aligner.query_end_gap_score)


alignments = aligner.align(expected[0].seq,expected[1].seq)
print("JELEN 5-6 = ",249-alignments[0].score, "Number of alignments: %d" % len(alignments))
alignments = aligner.align(expected[1].seq,expected[2].seq)
print("JELEN 6-7 = ",249-alignments[0].score, "Number of alignments: %d" % len(alignments))
alignments = aligner.align(expected[0].seq,expected[2].seq)
print("JELEN 5-7 = ",249-alignments[0].score, "Number of alignments: %d" % len(alignments))

data = list(SeqIO.parse(pathFASTQ, "fastq"))
print(len(data))

for i in range(3):
    readings=[0] * 32

    for sequence in data:
        alignments = aligner.align(expected[i].seq,sequence.seq)
        distance= int(min(len(expected[i].seq),len(sequence.seq))-alignments[0].score)
        for k in range(32):
            if distance <=k:
                readings[k]+=1
    print(readings)
    plt.subplots()
    plt.plot(readings)
```

```python
mat = []
for i in range(0, len(data)):
    row = []
    if(len(data[i].seq)>220 and len(data[i].seq)<296):
        for j in range(0, len(data)):
            if(len(data[j].seq)>220 and len(data[j].seq)<296):
                row.append(249-aligner.align(data[i].seq, data[j].seq).score)
        mat.append(row)
        
        
        
```

```python
df = DataFrame(np.zeros((len(mat)+3, len(mat)+3)))

for i in range(0, len(mat)):
    for j in range(i, len(mat[i])):
        df[i][j] = mat[i][j]
        df[j][i] = mat[i][j]
```

```python
k = 0
for i in range(0, len(data)):
    if(len(data[i].seq)>220 and len(data[i].seq)<296):
        df[k][len(mat)] = 249-aligner.align(expected[0].seq, data[i].seq).score
        df[k][len(mat)+1] = 249-aligner.align(expected[1].seq, data[i].seq).score
        df[k][len(mat)+2] = 249-aligner.align(expected[2].seq, data[i].seq).score
        df[len(mat)][k] = df[k][len(mat)]
        df[len(mat)+1][k] = df[k][len(mat)+1]
        df[len(mat)+2][k] = df[k][len(mat)+2]
        k+=1
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])

#print(df)

#Initialize the class object
kmeans = KMeans(n_clusters= 3)
 
#predict the labels of clusters.
label = kmeans.fit_predict(principalDf)

#Getting unique labels
 
u_labels = np.unique(label)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(principalComponents[label == i , 0] , principalComponents[label == i , 1] , label = i)

#plt.scatter(principalComponents[label == 146 , 0], principalComponents[label == 146 , 1], 'r')

df1 = pd.DataFrame(data=principalDf.tail(3))
for i in range(0,len(df1)):
    plt.scatter(df1.iloc[i]['PC1'], df1.iloc[i]['PC2'], color='red')
    #plt.scatter(i[0], i[1], 'r')



plt.legend()
plt.show()
```

```python
expected = list(SeqIO.parse(path, "fasta"))
    
from Bio import Align
aligner = Align.PairwiseAligner()
aligner.mode = 'local'

aligner.match_score = 1
#aligner.mismatch_score = 0.0
aligner.open_gap_score = -1
aligner.extend_gap_score = -1
aligner.target_end_gap_score = 0
aligner.query_end_gap_score = -0.5

#print(aligner.query_end_gap_score)


alignments = aligner.align(expected[0].seq,expected[1].seq)
print("JELEN 5-6 = ",249-alignments[0].score, "Number of alignments: %d" % len(alignments))
alignments = aligner.align(expected[1].seq,expected[2].seq)
print("JELEN 6-7 = ",249-alignments[0].score, "Number of alignments: %d" % len(alignments))
alignments = aligner.align(expected[0].seq,expected[2].seq)
print("JELEN 5-7 = ",249-alignments[0].score, "Number of alignments: %d" % len(alignments))

data = list(SeqIO.parse(pathFASTQ, "fastq"))

for i in range(3):
    readings=[0] * 32

    for sequence in data:
        alignments = aligner.align(expected[i].seq,sequence.seq)
        distance= int(min(len(expected[i].seq),len(sequence.seq))-alignments[0].score)
        for k in range(32):
            if distance <=k:
                readings[k]+=1
    print(readings)
    plt.subplots()
    plt.plot(readings)
```

```python
def avgDistances(expected, sequence):
    distances = {}
    for exp in expected:
        dictionary = {}
        sumB = 0
        sumN = 0
        for seq in sequence:
            alignments = aligner.align(exp.seq, seq.seq)
            distance= int(min(len(exp.seq),len(seq.seq))-alignments[0].score)
            if(distance not in dictionary and abs(distance) < 2* len(exp.seq)):
                dictionary[distance] = 0
            if(abs(distance) < 2* len(exp.seq)):
                dictionary[distance]+=1
        for key in sorted(dictionary):
            value = dictionary[key]
            sumB += sumB+abs(key*value)
            sumN += value
            print(sumB, abs(key*value), sumN)
            
        distances[exp.id] = sumB/sumN
        print(distances, sumB, sumN)
    return distances
```

```python
path = os.path.join(os.path.dirname(notebook_path), "data/J29B_expected.fasta")
pathFASTQ= os.path.join(os.path.dirname(notebook_path), "data/fastq/J29_B_CE_IonXpress_005.fastq")

```

```python
directory = 'data/fastq/'
dict1 = {}

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f) and filename.startswith('J'):
        dataTest = list(SeqIO.parse(f, "fastq"))
        dict2 = avgDistances(expected, dataTest)
        for dt in dict2:
            if(dt not in dict1):
                dict1[dt] = []
            dict1[dt].append(dict2[dt])
```

```python
df= pd.DataFrame(dict1)
print(df)
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PC1', 'PC2'])
#print(principalDf)
```

```python
K = range(1,20)
Sum_of_squared_distance = []
for num_clusters in K :
 kmeans = KMeans(n_clusters=num_clusters)
 kmeans.fit(principalDf)
 Sum_of_squared_distance.append(kmeans.inertia_)
plt.plot(K,Sum_of_squared_distance,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Sum of squared distances/Inertia') 
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
#Initialize the class object
kmeans = KMeans(n_clusters= 6)
 
#predict the labels of clusters.
label = kmeans.fit_predict(principalDf)

#Getting unique labels
 
u_labels = np.unique(label)
 
#plotting the results:
 
for i in u_labels:
    plt.scatter(principalComponents[label == i , 0] , principalComponents[label == i , 1] , label = i)
plt.legend()
plt.show()
```
